```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Read the dataset
dataset = pd.read_csv('D:/recomendation/Crop_recommendation.csv')

# Split into features and labels
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train Random Forest model
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)

# --- Jupyter-friendly input ---
n_params = 80
p_params = 20
k_params = 43
t_params = 10
h_params = 70
ph_params = 6.5
r_params = 203

user_input = np.array([[n_params, p_params, k_params, t_params, h_params, ph_params, r_params]])

# Make prediction
predictions = classifier.predict(user_input)
print("Predicted Crop:", predictions[0])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import joblib # <-- 1. Import joblib
import os

# --- Setup ---
# Create a models directory if it doesn't exist
if not os.path.exists('models'):
    os.makedirs('models')

# --- Data Loading and Training ---
# Read the dataset
try:
    dataset = pd.read_csv('Crop_recommendation.csv')
except FileNotFoundError:
    print("Error: 'Crop_recommendation.csv' not found. Please place it in the same directory.")
    exit()


# Split into features and labels
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train Random Forest model
classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
print("Model training complete.")


# --- 2. Save the Trained Model ---
# Define the path and filename for the model
model_filename = 'models/crop_recommendation_model.joblib'

# Save the classifier object to a file
joblib.dump(classifier, model_filename)
print(f"Model saved successfully to {model_filename}")


# --- Example Prediction (Optional) ---
# You can keep this part to test the model before saving
n_params = 90
p_params = 42
k_params = 43
t_params = 21
h_params = 82
ph_params = 6.5
r_params = 203

user_input = np.array([[n_params, p_params, k_params, t_params, h_params, ph_params, r_params]])

predictions = classifier.predict(user_input)
print("Predicted Crop:", predictions[0])
```
